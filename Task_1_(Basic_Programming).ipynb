```python
# 5. Loops

n = int(input())

for i in range(n):
    print(i * i)
```

```python
# 6. Write A Function

def is_leap(year):
    leap = False

    # Write your logic here
    if year % 400 == 0:
        leap = True
    elif year % 100 == 0:
        leap = False
    elif year % 4 == 0:
        leap = True
    return leap

year = int(input())
print(is_leap(year))
```
