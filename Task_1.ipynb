## **Running Sum of 1d Array**

```python
class Solution:
    def runningSum(self, nums):
        run = 0
        ans = []
        for n in nums:
            run += n
            ans.append(run)
        return ans
```

## **Shuffle the Array**

```python
class Solution:
    def shuffle(self, nums, n):
        res = []
        for i in range(n):
            res.append(nums[i])
            res.append(nums[i+n])
        return res
```

## **Kids With the Greatest Number of Candies**

```python
class Solution:
    def kidsWithCandies(self, candies, extraCandies):
        m = max(candies)
        return [c + extraCandies >= m for c in candies]
```
